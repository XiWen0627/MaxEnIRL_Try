## 读取奖励文件

```python
data = pd.read_csv(dataFile, encoding='utf-8')
```

## 读取属性文件

```python
os.chdir(r'F:\BaiduNetdiskDownload\共享单车轨迹\共享单车轨迹\01_研究生毕业论文\1_2_DataPreProcess')
dataAttrFile = r'./stateAttrNorm.csv'
```

```python
dataAttr = pd.read_csv(dataAttrFile, sep=',')
```

## 数据预处理

```python
dataAttr1 = dataAttr[dataAttr['Pass']!=0]
dataAttr2 = dataAttr1.drop(['Pass'],axis=1)
dataMerge = pd.merge(data, dataAttr2,left_on='index',right_on='FID')
dataAttr3 = dataMerge.drop(['Unnamed: 0','FID'], axis=1)
dataAttr3
```

```python
mean = dataAttr3['reward'].mean()
std = dataAttr3['reward'].mean()
dataAttr3['reward_Norm'] = dataAttr3['reward'].apply(lambda x: (x-mean/std))
dataAttr3
```

```python
X = dataAttr3.drop(['index','reward','reward_Norm'],axis=1)
y = dataAttr3['reward_Norm']
```

## 利用可解释机器学习 SHAP 对模型训练结果进行解释

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, 
                              random_state=1, 
                              n_jobs=1)
model.fit(x_train, y_train)
```

```python
y_pre = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

r2 = r2_score(y_true=y_test, y_pred=y_pre)
```

```python
r2
```

## SHAP 解释结果可视化

```python
import shap
import matplotlib.pyplot as plt
```

### 蜂群图 Beeswarm Plot

```python
explainer1 = shap.Explainer(model, X)
shap_values1 = explainer1(X)
```

```python
shap.plots.beeswarm(shap_values1,color=plt.get_cmap('GnBu'), max_display=18)
```

### 对蜂群图的解释
- 影响要素
- 正向影响/负向影响
- 影响的相对大小

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
```

### 依赖图 Dependence Plot

```python
shap.dependence_plot("Car", shap_values, X, cmap='GnBu', interaction_index="Car")
```

- 街景中车辆要素的占比与其对奖励值的贡献基本呈现出负向的线性相关关系；

```python
shap.dependence_plot("Car", shap_values, X, cmap='GnBu')
```

- 现象：绿点大部分位于蓝点之上
- 也就是说，街景中车辆占比相同的情况下，低路灯占比对骑行奖励的贡献要高于高交通灯占比

```python
shap.dependence_plot("Car", shap_values, X, interaction_index="Vegetation", cmap='GnBu')
# shap.dependence_plot("TrafLight", shap_values, X, cmap='GnBu')
```

相互依赖关系存在变化：
- 蓝点在红点之上：在车辆要素占比较低时，对于相同的车辆要素占比，较低的绿视率会带来更高的奖励；
- 红点在蓝点之上：在车辆要素占比较高时，对于相同的车辆要素占比，较高的绿视率会带来更高的奖励。

```python
shap.dependence_plot("Vegetation", shap_values, X, interaction_index="Vegetation", cmap='GnBu')
```

```python
shap.dependence_plot("Vegetation", shap_values, X, interaction_index="Terrain", cmap='GnBu')
```

```python
shap.dependence_plot("Vegetation", shap_values, X, interaction_index="Slope", cmap="GnBu")
```

绿视率低：车辆少奖励高
绿视率高：车辆多奖励高

### 交互图 Interaction Plot

```python
shap_interaction_values = explainer.shap_interaction_values(X)
shap.summary_plot(shap_interaction_values, X, cmap='GnBu')
```

### SHAP Interaction Value Matrix

```python
interaction_matrix = np.abs(shap_interaction_values).sum(0)
for i in range(interaction_matrix.shape[0]):
    interaction_matrix[i, i] = 0
inds = np.argsort(-interaction_matrix.sum(0))[:14]    # 返回数组元素排序后的索引
sorted_ia_matrix = interaction_matrix[inds, :][:, inds]    # 对矩阵的数值进行排序
plt.figure(figsize=(6, 6))
plt.imshow(sorted_ia_matrix, cmap='GnBu')
plt.yticks(
    range(sorted_ia_matrix.shape[0]),
    X.columns[inds],
    rotation=50.4,
    horizontalalignment="right",
)
plt.xticks(
    range(sorted_ia_matrix.shape[0]),
    X.columns[inds],
    rotation=50.4,
    horizontalalignment="left",
)
plt.title("SHAP Interaction Values", fontsize=18)
plt.gca().xaxis.tick_top()
plt.show()
```

### SHAP Interaction Value Dependence Plots
- Running a dependence plot on the SHAP interaction values allows us to separately observe the main effects and the interaction effects.
- 排除与其他特征的交互作用，以得到净效应

```python
shap.dependence_plot(("Vegetation","Vegetation"), shap_interaction_values, X, cmap='GnBu_r')
```

```python
shap.dependence_plot(("Vegetation","Car"), shap_interaction_values, X, cmap='GnBu_r')
```

Vegetation 与 Car 交互作用对共享单车骑行奖励的贡献
