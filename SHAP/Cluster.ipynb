```python
def find_elbow_point(variance_ratio):
    for i in range(1, len(variance_ratio)):
        slope = variance_ratio[i] - variance_ratio[i - 1]
        if slope < 0.05:  # 调整斜率的阈值以寻找肘部点
            return i
```

```python
elbow_point = find_elbow_point(y2)
print(f"Elbow point is at {elbow_point} components")
```

#### 使用 K-Means 聚类算法

```python
from sklearn.cluster import KMeans
```

##### 从结果来看 3类 结果相对来讲比较理想

```python
sh_scores = []
ch_scores = []
DB_scores = []

for i in range(8):
    n_cluster = int(i+2)
    kmeans = KMeans(n_clusters=n_cluster)
    kmeans.fit_predict(x_new)
    labels = kmeans.labels_
    # 观感上来讲, 倘若同属一个聚类的点样本数目非常大轮廓系数的表现可能会不错
    sh_score = metrics.silhouette_score(x_new, labels)    # 轮廓系数, 每一个点与同一聚类中其他点的相似程度
    # Calinski-Harabasz Criterion, 聚类结果具有较大簇间方差与较小的簇内方差 - The Smaller, the greater
    ch_score = metrics.calinski_harabasz_score(x_new, labels)    
    # Davies-Bouldin Criterion, 聚类内部与聚类间距离的比例 - The smaller, the greater
    DB_score = metrics.davies_bouldin_score(x_new, labels)
    
    sh_scores.append(sh_score)
    ch_scores.append(ch_score)
    DB_scores.append(DB_score)
    
print('轮廓系数的变化为{}'.format(sh_scores))
print('CH值的变化为{}'.format(ch_scores))
print('DB值的变化为{}'.format(DB_scores))
```

#### 使用 Spectral Clustering

##### 谱聚类的结果是将降维后的数据划分为三种典型的偏好

```python
from sklearn.cluster import SpectralClustering
```

```python
spclustering = SpectralClustering(n_clusters=6, 
                                  affinity='nearest_neighbors')
spclustering.fit_predict(x_new)
labels_sp = spclustering.labels_
```

```python
sh_scores_sp = []
ch_scores_sp = []
DB_scores_sp = []

for i in range(8):
    n_cluster = int(i+2)    # 从 2 开始
    spclustering = SpectralClustering(n_clusters=n_cluster, 
                                  affinity='nearest_neighbors')
    spclustering.fit_predict(x_new)
    labels = spclustering.labels_
    # 观感上来讲, 倘若同属一个聚类的点样本数目非常大轮廓系数的表现可能会不错
    sh_score = metrics.silhouette_score(x_new, labels)    # 轮廓系数, 每一个点与同一聚类中其他点的相似程度
    # Calinski-Harabasz Criterion, 聚类结果具有较大簇间方差与较小的簇内方差 - The Bigger, the greater
    ch_score = metrics.calinski_harabasz_score(x_new, labels)    
    # Davies-Bouldin Criterion, 聚类内部与聚类间距离的比例 - The bigger, the greater
    DB_score = metrics.davies_bouldin_score(x_new, labels)
    
    sh_scores_sp.append(sh_score)
    ch_scores_sp.append(ch_score)
    DB_scores_sp.append(DB_score)
    
print('轮廓系数的变化为{}'.format(sh_scores_sp))
print('CH值的变化为{}'.format(ch_scores_sp))
print('DB值的变化为{}'.format(DB_scores_sp))
```
